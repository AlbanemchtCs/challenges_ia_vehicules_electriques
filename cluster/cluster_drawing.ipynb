```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
Data = pd.read_csv('C:/Users/Admin/Downloads/data_commune_cluster.csv')
Data.columns
```

```python
Data[['population','com_life_level','rurality']].hist()
```

```python
plt.scatter(Data['longitude'],Data['latitude'],c=Data['cluster'],s=Data['population']/10000, cmap = 'tab20c')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.title('clusters des communes')
```

```python
filt  = Data[np.sqrt((Data['latitude']-float(Data[Data['codgeo']== '75056']['latitude']))**2+
             (Data['longitude']-float(Data[Data['codgeo']== '75056']['longitude']))**2) < 15000]

fig, ax = plt.subplots()
ax.scatter(filt['longitude'],filt['latitude'],c=filt['cluster'],s=filt['population']/700, cmap = 'tab20c')
for index,i in filt.iterrows():
    ax.text(i['longitude'],i['latitude'],i['libgeo'],size=np.log(i['population']/10),horizontalalignment='center',
     verticalalignment='center')
```
